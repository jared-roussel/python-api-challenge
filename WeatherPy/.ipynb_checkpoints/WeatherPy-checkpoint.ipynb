```python
#Import dependences
import pandas as pd
import json
import matplotlib.pyplot as plt
import random
from citipy import citipy
import numpy as np
import requests
from pprint import pprint
#from config import weather_api_key
weather_api_key = '877069c5920cb3ad3f4efa6dac5c6f9e'
```

## Generate Cities List

```python
#Step1: Need random latitude and longitude values to feed to citipy

#Create lists and add random latitude and longitude values
long_list = [random_long for random_long in np.random.randint(-180,180,500)]
lat_list = [random_lat for random_lat in np.random.randint(-90,90,500)]


#Feed random latitude and longitude values to citipy
cities = []
index_max = len(lat_list)
#index_max
for index in range(0,index_max):
    lat_value = lat_list[index]
    #print(lat_value)
    long_value = long_list[index]
    #print(long_value)
    city = citipy.nearest_city(lat_value,long_value)
    city_name = city.city_name
    cities.append(city_name)
print(cities)



```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#api.openweathermap.org/data/2.5/weather?q={city name}&appid={API key}

lat_list_api = []
long_list_api = []
max_temp_list = []
humidity_list = []
cloudiness_list = []
wind_speed_list = []
country_list = []
date_list = []

base_url = "https://api.openweathermap.org/data/2.5/weather?q="

for city_search in cities:
    city_index = cities.index(city_search)
    #city_index
    print(f'City number {city_index}, the city of {city_search} is being processed.')
    print('-----------------------------------------------------------------')
    query_url = f'{base_url}{city_search}&appid={weather_api_key}'
    response = requests.get(query_url).json()
    #pprint(response)
    
    try:
        lat = response['coord']['lat']
        lat_list_api.append(lat)

        long = response['coord']['lon']
        long_list_api.append(long)

        max_temp = response['main']['temp_max']
        max_temp_list.append(max_temp)

        humidity = response['main']['humidity']
        humidity_list.append(humidity)

        cloudiness = response['clouds']['all']
        cloudiness_list.append(cloudiness)

        wind_speed = response['wind']['speed']
        wind_speed_list.append(wind_speed)

        country = response['sys']['country']
        country_list.append(country)

        date = response['dt']
        date_list.append(date)
    
    except:
        print(f'City {city_search} is not found.')
        print('-----------------------------------------------------------------')
    
    
    

   



```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_info_dict = {
    "City": city_search,
    "Lat": lat_list_api,
    "Long": long_list_api,
    "Max Temp": max_temp_list,
    "Humidity": humidity_list,
    "Cloudiness": cloudiness_list,
    "Wind Speed": wind_speed_list,
    "Country": country_list,
    "Date": date_list
}

city_info_df = pd.DataFrame(city_info_dict)
city_info_df.head()
```
